```python
# Importing libs
import pandas as pd
import seaborn as srn
#seaborn can create more sophisticated boxplots
```

```python
# Importing dataset
df = pd.read_csv('datasets/trees.csv')
df.head()
```

```python
# Visualizing boxplot
srn.boxplot(data=df.Volume, orient='v').set_title('Trees') 
# data= your variable, orient = 'v' for vertical and 'h' for horizontal, set_title('boxplot title')
```

```python
# Visualizing a lot of boxplots at the same image
srn.boxplot(data = df)
```
