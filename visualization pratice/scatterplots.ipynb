```python
# Scatterplotss graphics are excellent for comparing two numerical variables
```

```python
# Importing libs
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing dataset
df = pd.read_csv('trees.csv')
df.head()
```

```python
# Scatterplots graphic considering volume and Scatterplots
plt.scatter(x=df.Girth, y=df.Volume, color = 'blue', facecolors = 'none', marker = '*')
plt.title('Trees')
plt.xlabel('Circumference')
plt.ylabel('Volume')
```

```python
# As we can see, when the circumference increases the volume increases too
```

```python
# Line graphic considering volume and "girth" - no marker
plt.plot(df.Girth,df.Volume)
plt.title('Trees')
plt.xlabel('Circumference')
plt.ylabel('Volume')
```

```python
# Scatterplots graphic with data "distance" (jitter)
#fit_reg tendency line
sns.regplot(x=df.Girth,y=df.Volume, data = df, x_jitter = 0.3, fit_reg = False)
```

```python
# One issue of Scatterplots graphics is that it can overlap data, one way to counter this is using jitter, he adds randomness to the data when lots of points are in the same place
```
