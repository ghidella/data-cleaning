```python
# Importing libs 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing dataset
df = pd.read_csv('datasets/trees.csv')
df.head()
```

```python
# Histogram with 6 divisions (bins), only for the first attribute of the database
plt.hist(df.iloc[:,1], bins = 6)
```

```python
# Histogram with frequency distribution line, 6 divisions(bins)
#kde = density line
sns.histplot(df.iloc[:,1], kde=False, bins=6, color='blue').set(title='Trees')
```

```python
# Density, kdeplot method for density
sns.kdeplot(df.iloc[:,1], color='blue').set(title='Trees')
```

```python
# Density and histogram
sns.histplot(df.iloc[:,1], kde=True, bins=6, color='blue').set(title='Trees')
```
