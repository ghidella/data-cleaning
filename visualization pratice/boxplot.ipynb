```python
# Importing libs 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing dataset
df = pd.read_csv('trees.csv')
df.head()
```

```python
# Generating boxplot
plt.boxplot(df.Volume, vert = False, showfliers = False, notch = True, patch_artist = True)
# vert = False -> horizontal boxplot, showfliers -> outliers, notch = True -> notches the median, patch_artist -> fills the boxplot
plt.title('Trees')
plt.xlabel('Volume')
```

```python
# Data by column
plt.boxplot(df)
plt.title('Trees')
plt.xlabel('Volume')
```

```python
# Generating 3 boxplots with different information each
plt.boxplot(df.Volume, vert = False)
plt.boxplot(df.Girth, vert = False)
plt.boxplot(df.Height, vert = False)
plt.title('Trees')
plt.xlabel('Volume')
```
