```python
# Importing libs
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
# Importing dataset
df = pd.read_csv('datasets/trees.csv')
df.head()
```

```python
# Histogram with 10 divisions (bins) and density graphic (kernel density - smoothing)
srn.histplot(df.Volume, color="blue", kde=True, stat="density", bins=10).set(title='Volume')
# df.variable, stat: (count, frequency, probability, percent, density)
```
